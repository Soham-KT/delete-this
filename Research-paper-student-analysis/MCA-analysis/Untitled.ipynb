```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import scipy.cluster.hierarchy as sch
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
%matplotlib inline
```

```python
def find_num_neighbours(data, max_n):
    means = []
    inertia = []
    sill_score = []

    for i in range(2, max_n):
        kmeans = KMeans(n_clusters=i)
        kmeans.fit(data)
        sill_score.append(silhouette_score(df, kmeans.fit_predict(df)))
        means.append(i)
        inertia.append(kmeans.inertia_)
        

    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 7))
    
    ax[0].plot(means, inertia, 'o-')
    ax[0].set_xlabel('Number of clusters')
    ax[0].set_ylabel('Cost')
    ax[0].grid(True)
    
    ax[1].plot(means, sill_score, 'o-')
    ax[1].set_xlabel('Number of clusters')
    ax[1].set_ylabel('shilhouette-score')
    ax[1].grid(True)
```

```python
df = df.fillna(0)
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import normalize

df = pd.DataFrame(normalize(df)) 
df.head()
```

```python
find_num_neighbours(df, 20)
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(df)
pred_kmeans = kmeans.predict(df)
print(f'Shilhouette-score: {silhouette_score(df, pred_kmeans)}')
```

```python
dendrogram = sch.dendrogram(sch.linkage(df, method='ward'))
```

```python
hier = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
pred_hier = hier.fit_predict(df)
print(f'Shilhouette-score: {silhouette_score(df, pred_hier)}')
```
