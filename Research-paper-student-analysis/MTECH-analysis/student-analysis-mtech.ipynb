```python
def bin_cgpa(x: int) -> None: 
    if (x - int(x)) > 0.75:
        x = int(x) + 1

    elif (x - int(x)) <= 0.75 and (x - int(x)) > 0.25:
        x = int(x) + 0.5
        
    else:
        x = int(x)
    
    return x
```

```python
df['CGPA-bin'] = df['CGPA-bin'].apply(bin_cgpa)
```

# Clustering

## clustering on : prior knowledge

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X = df.drop(['prior knowledge (1-5)'], axis=1)
Y = df['prior knowledge (1-5)']

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(X)

Y = Y.values.reshape(-1, 1)
y_scaled = scaler.fit_transform(Y)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, shuffle=True)

knc = KNeighborsRegressor()
knc.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_pred = knc.predict(x_test)
print(f"Mean absolute error: { mean_absolute_error(y_test, y_pred) }")
print(f"Mean squared error: { mean_squared_error(y_test, y_pred) }")
```

```python
y_all = knc.predict(X)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
def find_num_neighbours(data, max_n):
    means = []
    inertia = []
    sill_score = []

    for i in range(2, max_n):
        kmeans = KMeans(n_clusters=i)
        kmeans.fit(data)
        sill_score.append(silhouette_score(df, kmeans.fit_predict(df)))
        means.append(i)
        inertia.append(kmeans.inertia_)
        

    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 7))
    
    ax[0].plot(means, inertia, 'o-')
    ax[0].set_xlabel('Number of clusters')
    ax[0].set_ylabel('Cost')
    ax[0].grid(True)
    
    ax[1].plot(means, sill_score, 'o-')
    ax[1].set_xlabel('Number of clusters')
    ax[1].set_ylabel('shilhouette-score')
    ax[1].grid(True)
```

```python
find_num_neighbours(df, 20)
```

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(df)
```

```python
print(f'Silhouette score : {silhouette_score(df, kmeans.fit_predict(df))}')
```
